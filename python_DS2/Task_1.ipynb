### Урок 1. Введение в задачу классификации. Постановка задачи и подготовка данных.

----
#### Задание 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Максимизация Precision была бы полезна в удостоверении личности, например при проверки соответсвия отпечатка пальца или фотографии.
А также при принятии решения о благоприятных для запуска космического корабля условий (благоприятные условия для пуска или нет).


Максимизация Recall была бы кстати в процессах контроля качества на производстве.
Например, при определении бракованных деателей при производстве частей автомобиля.
А также в системах аварийного срабатывания.
Например, пожарной сигнализации и т.п.

----
#### Задание 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Фактически нас интересует минимизация ошибок разного рода и фактически Precision и Recall это величины обратные разным видам ошибок (числители у них одинаковые, а вот в знаменателях используется количество разных видов ошибок). 
В таких случаях среднее арифмитическое будет неверным, для обратных величин надо использовать среднее гармоническое.
