# Урок 2. Анализ данных и проверка статистических гипотез.

----
### Задание 1.

*В чём различие между зависимыми и независимыми выборками?*

----

Зависимые выборки это выборки при которых важна связь между выборками (например, мы хотим изучить влияние какого-то события на человека и опрашиваем его "до и после", при этом нам важна связь между выборками, т.е. нам важно связать ответы конкретного человека "до" и его же ответы "после").

В независимых выборках никакие связи нам не важны.

Относительно анализа данных и проверки стат. гипотез зависимость/назваисимость выборок влияет на инструменты, которые мы можем использовать.

Например, Критерий хи-квадрат можно использовать только для независимых выборок.

----

### Задание 2.

*Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?*

----

Параметрические критерии применяются если величина распределена "нормально".

Если она распределена "не нормально" испольщуют непараметрический аналог.
